#### Rebalancing portfolios
In this exercise, you are asked to look at the effects of periodically rebalancing portfolios. We define rebalancing here to mean an adjustment to the portfolio weights back to their target weights after differences in returns achieved by different assets have led to deviations from those weights. In the example studied in this problem set, the target portfolio will be equal-weighted, i.e., the target weight of each asset in a portfolio consisting of $N$ assets will be $\frac{1}{N}$.

1. To prepare the subsequent analyses, please create simulated returns as follows. There are 100 assets and 10000 (corresponding to about 40 years) logarithmic daily returns for each of these assets. Returns follow a multivariate normal distribution with expected log returns of 0.05% and a daily standard deviation of log returns of 2% (which is similar to historical real stock returns; thus we will henceforth refer to the assets as stocks). We will assume that pairwise correlations of returns between different stocks are identical for all pairs of stocks. Please create one numpy array each for pairwise correlations equal to 0.5 and 0.2. In subsequent questions, you are expected to always use each of these arrays, which allows you to compare results later.

```python
import numpy as np 
import matplotlib.pyplot as plt
#initialisation - set basic values
rng = np.random.default_rng(12345) 
mean = 0.0005
std_dev = 0.02 
num_assets = 100
num_returns = 10000 
yrs = 40
cor = np.identity(100) #create an identity matrix, this will be used to create the correlation matrix
corr_05 = np.full((num_assets,num_assets),0.5) #correlation matrix for corr 0.5
np.fill_diagonal(corr_05, 1)
corr_02 = np.full((num_assets,num_assets),0.2) #correlation matrix for corr 0.2
np.fill_diagonal(corr_02, 1)
```

2. For each of these arrays, plot the return time series generate above for five stocks randomly chosen among the 100 stocks contained in the array.

```python
#we creating returns, i was getting an error when i tried to use the variable mean, so lets rescale mean to fit the dimensions of returns matrices
mean_vector = np.full(num_assets,mean)
#Array 1 

#create a covariance matrix
cov_05 = corr_05*(std_dev**2)
#generate returns
returns05 = rng.multivariate_normal(mean = mean_vector, cov = cov_05, size = num_returns) 
#randomly choose 5 stocks
random_stocks_05 = rng.choice(range(num_assets),5, replace = False)
plot05 = returns05[:, random_stocks_05]
#plot the returns
plt.figure(figsize=(12, 6))
graph = plt.plot(plot05)  # Plot only the first 1000 days for clearer visibility
labels = [f'Stock {i+1}' for i in range(5)]
plt.legend(graph, labels, fontsize=12)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Return', fontsize=18)
plt.title('Return Time Series for 5 Randomly Selected Stocks (pairwise correlations of 0.5)')
plt.show()

#Array 2 - analogically to array 1
#create a covariance matrix
cov_02 = corr_02*(std_dev**2)
#generate returns
returns02 = rng.multivariate_normal(mean = mean_vector, cov = cov_02, size = num_returns) 
#randomly choose 5 stocks 
random_stocks_02 = rng.choice(range(num_assets),5, replace = False)
plot02 = returns02[:, random_stocks_05]
#plot the returns
plt.figure(figsize=(12, 6))
graph = plt.plot(plot02)  
labels = [f'Stock {i+1}' for i in range(5)]
plt.legend(graph, labels, fontsize=12)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Return', fontsize=18)
plt.title('Return Time Series for 5 Randomly Selected Stocks (pairwise correlations of 0.2)')
plt.show()
```

3. Compute the cumulative returns for each array and plot cumulative returns for the same stocks as in the previous question.

```python
#Array 1 
cumul_return05 = np.cumsum(returns05[:, random_stocks_05], axis = 0) #compute the comulative returns for the random_stocks_05
plt.figure(figsize=(12, 6))
graph = plt.plot(cumul_return05)  
labels = [f'Stock {i+1}' for i in range(5)]
plt.legend(graph, labels, fontsize=12)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Return', fontsize=18)
plt.title('Time Series of cumulative returns for 5 Randomly Selected Stocks (pairwise correlations of 0.5)')
plt.show()

#Array 2 - analogically to array 1
cumul_return02 = np.cumsum(returns02[:, random_stocks_02], axis = 0) 
plt.figure(figsize=(12, 6))
graph = plt.plot(cumul_return02)  
labels = [f'Stock {i+1}' for i in range(5)]
plt.legend(graph, labels, fontsize=12)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Return', fontsize=18)
plt.title('Time Series of cumulative returns for 5 Randomly Selected Stocks (pairwise correlations of 0.2)')
plt.show()
```

## Daily rebalancing
### Corr 05
#### 2 stock portfolio
---

```python
#compounding is done similarly to the case with no rebalancing, but after rebalancing we store the rebalanced values on the day i to be the starting values
#for the day i+1
#initialisation - set initial values after investing 1000 000 
initial_starting_values_2 = np.full((100,2), 500000)
starting_values_2 = np.full((100,2), 500000)
initial_starting_values_10 = np.full((100,10), 100000)
starting_values_10 = np.full((100,10), 100000)
for i in range(10000-1):
   # compute the stock prices after the ith day
    returns_day_i = np.array([returns05[i, stocks] + 1 for stocks in portfolio_2])
    values_after_day_i = returns_day_i * starting_values_2
    #rebalance 
    total_values = np.sum(values_after_day_i, axis = 1)
    rebalanced_values_2_05 = np.repeat(total_values/2, 2).reshape(100,2)
    starting_values_2 = rebalanced_values_2_05

#store final values compute statistics 
final_values_daily_2_05 = rebalanced_values_2_05
final_returns_daily_2_05 = np.log((final_values_daily_2_05 / initial_starting_values_2))
avg_log_rebalanced_returns_daily_2_05 = np.mean(final_returns_daily_2_05, axis=1)
mean_final_values_daily_2_05 = np.mean(final_values_daily_2_05)
std_final_values_daily_2_05 = np.std(final_values_daily_2_05)
```

```python
print(f"Final mean value in the case of 2 stocks portfolio and corr 0.5 with no rebalancing is {mean_final_values2_05:,.2f}")
print(f"Final mean value in the case of 2 stocks portfolio and corr 0.5 with daily rebalancing is {mean_final_values_daily_2_05:,.2f}")
print(f"Final mean value in the case of 2 stocks portfolio and corr 0.5 with 50-day rebalancing is {mean_final_values_50days_2_05:,.2f}")


print(f"Mean std_dev in the case of 2 stocks portfolio and corr 0.5 with no rebalancing is {std_final_values2_05:,.2f}")
print(f"Mean std_dev in the case of 2 stocks portfolio and corr 0.5 with daily rebalancing is {std_final_values_daily_2_05:,.2f}")
print(f"Mean std_dev in the case of 2 stocks portfolio and corr 0.5 with 50-day rebalancing is {std_final_values_50days_2_05:,.2f}")



print(f"Final mean value in the case of 2 stocks portfolio and corr 0.2 with no rebalancing is {mean_final_values2_02:,.2f}")
print(f"Final mean value in the case of 2 stocks portfolio and corr 0.2 with daily rebalancing is {mean_final_values_daily_2_02:,.2f}")
print(f"Final mean value in the case of 2 stocks portfolio and corr 0.2 with 50-day rebalancing is {mean_final_values_50days_2_02:,.2f}")


print(f"Mean std_dev in the case of 2 stocks portfolio and corr 0.2 with no rebalancing is {std_final_values2_02:,.2f}")
print(f"Mean std_dev in the case of 2 stocks portfolio and corr 0.2 with daily rebalancing is {std_final_values_daily_2_02:,.2f}")
print(f"Mean std_dev in the case of 2 stocks portfolio and corr 0.2 with 50-day rebalancing is {std_final_values_50days_2_02:,.2f}")

```

## Comments 
---
It seems that in the case of 2 stocks portfolio and correlation of 0.5 we can achieve slightly  higher mean final value with daily rebalancing, rebalancing every 50 days lowers our mean final value. We get the lowest standard deviation with 50 days rebalancing.We are getting reasonably higher values for the case with correlation of 0.2. We can get higher mean final values with rebalancing every 50 days in this case, the effect of rebalancing is here much higher. However, we are not able to dampen the volatility with lower correlation even with rebalancing. It is actually slighlt higher with rebalancing. It seems that if the portfolio is less diversified (higher correlation) than rebalancing acts to decreaes the volatility, but with more diversified portfolio we can get considerably great mean final values with rebalancing.

```python
print(f"Final mean value in the case of 10 stocks portfolio and corr 0.5 with no rebalancing is {mean_final_values10_05:,.2f}")
print(f"Final mean value in the case of 10 stocks portfolio and corr 0.5 with daily rebalancing is {mean_final_values_daily_10_05:,.2f}")
print(f"Final mean value in the case of 10 stocks portfolio and corr 0.5 with 50-day rebalancing is {mean_final_values_50days_10_05:,.2f}")

#It is clear that not rebalancing the portfolios leads to much higher final values. Rebalancing daily leads to lower values than rebalancing every 50 days.


print(f"Mean std_dev in the case of 10 stocks portfolio and corr 0.5 with no rebalancing is {std_final_values10_05:,.2f}")
print(f"Mean std_dev in the case of 10 stocks portfolio and corr 0.5 with daily rebalancing is {std_final_values_daily_10_05:,.2f}")
print(f"Mean std_dev in the case of 10 stocks portfolio and corr 0.5 with 50-day rebalancing is {std_final_values_50days_10_05:,.2f}")

#We see that rebalancing the portfolios leads to less volatile final values, the lowest std dev is achieved with daily rebalancing


print(f"Final mean value in the case of 10 stocks portfolio and corr 0.2 with no rebalancing is {mean_final_values10_02:,.2f}")
print(f"Final mean value in the case of 10 stocks portfolio and corr 0.2 with daily rebalancing is {mean_final_values_daily_10_02:,.2f}")
print(f"Final mean value in the case of 10 stocks portfolio and corr 0.2 with 50-day rebalancing is {mean_final_values_50days_10_02:,.2f}")

#interestingly enough we can reach much higher mean final values with correlation of 0.2 than 0.5. Highest mean value is again achieved with no rebalancing. Rebalancing daily leads to slightly lower mean final values.

print(f"Mean std_dev in the case of 10 stocks portfolio and corr 0.2 with no rebalancing is {std_final_values10_02:,.2f}")
print(f"Mean std_dev in the case of 10 stocks portfolio and corr 0.2 with daily rebalancing is {std_final_values_daily_10_02:,.2f}")
print(f"Mean std_dev in the case of 10 stocks portfolio and corr 0.2 with 50-day rebalancing is {std_final_values_50days_10_02:,.2f}")

#we are also gettin much higher variance with 0.2 correlation than with 0.5 correlation. In terms of volatility it would be also advantageous to rebalance daily
```

We e are overall getting lower final values with 10 stock portfolios than in the case of 2 stock porfolio. Conversly, rebalancing has the greatest effect with lower correlation, but in this case we are also able to lower the volatility of the portfolios with rebalancing even with lower correlation. Analogically, we can significantly increase our final mean value with rebalancing in the case of lower correlation.
